````python
import pygraphviz
import networkx as nx
from networkx.drawing.nx_agraph import read_dot
import matplotlib.pyplot as plt

G = read_dot("../../test/bigf/Spacers_10853_GCAGACACGTTCCATTGGCTAATGTCGCCATTAAAGGCA_spacers.gv")
plt.figure(figsize=(10,10), facecolor="w", frameon=False)
pos=nx.spring_layout(G)
offset = .05
pos_labels = {}
keys = pos.keys()
for key in keys:
    x, y = pos[key]
    pos_labels[key] = (x, y+offset)

node_colors = [ G.node[key]['fillcolor'] for key in G.node ]
print(node_colors)
nx.draw_networkx_nodes(G, pos, node_size=1200, node_shape='o', node_color=node_colors)
nx.draw_networkx_edges(G, pos, width=2, edge_color='b')
nx.draw_networkx_labels(G, pos_labels, fontsize=2)
print (G.nodes())
plt.show()
````


> ```
Each spacer is coloured based on the coverage in a scale from blue to red. They are labelled using a four field scheme, where each field is separated by an "_" character. The fields are: nodeType_ID_cov_segment. The nodeType will be either sp or fl for spacer or flanker. The next three fields are the unique identification number, coverage of each flanker or spacer followed by the contig (linear segment) that the node was found in.
```

NetworkX/Matplotlib has issues with displaying nice graphs so to make the graph plots better I will try to use a module called pydot, which is a wraper to the graphviz layout programs.

````python
"""
pip3.5 install pydot2
pip3.5 install Pillow
"""

import pydot
from PIL import Image
from networkx.drawing import nx_pydot

Gp = nx_pydot.to_pydot(G)   #converts G to Gp in pydot format
edges = Gp.get_edge_list()
for e in edges:
    source = e.get_source()
    target = e.get_destination()
    if source == 'some string that is the name of the source with edges that you want to make look different' and target == 'some string2':
        e.set_style('bold')
        e.set_color('red')
        e.set_arrowhead('tee')  # use tee instead of arrow
        e.set_label('label for edge')

# now output your graph to a file and display it
outstem = 'graph'
Gp.write_png(outstem + '_dot.png', prog='dot')  # writes Gp to png file #use prog='neato' or 'fdp' for undirected graph (no default arrows for this)
# the next 2 lines open and display the png file
im = Image.open(outstem + '_dot.png') 
#im.show()
#from matplotlib.pyplot import imshow
import numpy as np
#imshow(np.asarray(im))
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10), facecolor="w", frameon=False)
plt.imshow(np.asarray(im))
plt.show()
````
